```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
import numpy as np
import warnings 
import warnings
warnings.filterwarnings('ignore')
```

```python
plt.figure(figsize=(10,10))
sb.countplot(data=df,x='winner',palette='icefire')
plt.title('MOST NUMBER OF WINS BY TEAM',color='BLUE', fontsize=20,fontname='Arial')
plt.xlabel('Winner',color='RED', fontsize=20,fontname='Arial')
plt.ylabel('Number of wins',color='RED', fontsize=20,fontname='Arial')
plt.show()
```

```python
sb.countplot(data=df,y='venue_name')
plt.title("MOST MATCHES IN STADIUM",color='PURPLE',fontsize=20)
plt.xlabel("Number of matches",color='Red')
plt.ylabel('Venue names')
plt.show()
```

```python
df['toss_won'].value_counts().plot(kind='pie',autopct='%.2f%%',shadow=True)

plt.title("TOSSES WIN PERCENTAGE BY TEAM",color='PURPLE',fontsize=20)

plt.ylabel('Toss_won',color='Red')
plt.show()
```

```python
df['decision'].value_counts().plot(kind='pie',autopct='%.2f%%',shadow=True)
plt.title("TOSS DECISION BY TEAM",color='PURPLE',fontsize=20)
plt.legend()
```

```python

import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrame

ipl_2023 = pd.DataFrame(df)

# Calculate the number of matches won by chasing and defending
matches_won_chasing = len(ipl_2023[ipl_2023['decision'] == 'BOWL FIRST'])
matches_won_defending = len(ipl_2023[ipl_2023['decision'] == 'BAT FIRST'])
labels = ['Chasing', 'Defending']
sizes = [matches_won_chasing, matches_won_defending]
plt.pie(sizes,autopct='%.1f%%', shadow=True, startangle=140)



plt.title('MATCHES WON BY CHASING AND DEFENDING',color='PURPLE')
plt.legend(labels)
plt.show()
```
